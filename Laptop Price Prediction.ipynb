```python
ssd_0 = []
hdd_0 = []
flash_storage_0 = []
hybrid_0 = []

for i in memory_cat[0]:
    if 'SSD' in i : 
        ssd_0.append(int(i.split()[0]))
    else : 
        ssd_0.append(0)
    if 'HDD' in i : 
        hdd_0.append(int(i.split()[0])) 
    else : 
        hdd_0.append(0)
    if 'Flash' in i : 
        flash_storage_0.append(int(i.split()[0]))
    else : 
        flash_storage_0.append(0)
    if 'Hybrid' in i : 
        hybrid_0.append(int(i.split()[0]))
    else : 
        hybrid_0.append(0)
```

```python
ssd_1 = []
hdd_1 = []
flash_storage_1 = []
hybrid_1 = []

for i in memory_cat[1]:
    if 'SSD' in i : 
        ssd_1.append(int(i.split()[0]))
    else : 
        ssd_1.append(0)
    if 'HDD' in i : 
        hdd_1.append(int(i.split()[0])) 
    else : 
        hdd_1.append(0)
    if 'Flash' in i : 
        flash_storage_1.append(int(i.split()[0]))
    else : 
        flash_storage_1.append(0)
    if 'Hybrid' in i : 
        hybrid_1.append(int(i.split()[0]))
    else : 
        hybrid_1.append(0)
```
